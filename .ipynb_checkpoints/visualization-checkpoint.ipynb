## Gender Distribution
There are multiple instances where a patient have multiple radiographs in the dataset. To separate these data, the *Image Index* column could be of help where the numbers separated by an underscore indicates a unique patient in the dataset.

```python
def get_patient_number(row):
    return row['Image Index'].split('_')[0]

# Find unique Image Indexes
chest_metadata['Patient Number'] = chest_metadata.apply(get_patient_number, axis=1)
```

```python
unique_patients = chest_metadata.drop_duplicates('Patient Number')

unique_patients
```

```python
plt.pie(unique_patients['Patient Gender'].value_counts(), labels = ['Male', 'Female'], autopct='%1.f%%')
plt.title('Gender Distribution', fontsize = 14)
plt.legend()
plt.savefig('images_made/GenderPieChart.png', dpi = 300, bbox_inches = 'tight')
```

# Findings based on Age Distribution
This section looks at if age is a factor to find a diagnosis on the patient.

The age would be split into these following sections:
- 0-12
- 13-17
- 18-30
- 31-45
- 46-59
- 60+

Afterwards, we would split the Finding Labels into two: Ones that have No Finding's and ones that are not 'No Finding'

```python
chest_metadata.loc[chest_metadata['Patient Age'] < 13, 'Patient Age Group'] = 'Children'
chest_metadata.loc[(chest_metadata['Patient Age'] >= 13) & (chest_metadata['Patient Age'] < 18), 'Patient Age Group'] = "Teens"
chest_metadata.loc[(chest_metadata['Patient Age'] >= 18) & (chest_metadata['Patient Age'] < 31), 'Patient Age Group'] = "Early Adults"
chest_metadata.loc[(chest_metadata['Patient Age'] >= 31) & (chest_metadata['Patient Age'] < 46), 'Patient Age Group'] = "Adults"
chest_metadata.loc[(chest_metadata['Patient Age'] >= 46) & (chest_metadata['Patient Age'] < 60), 'Patient Age Group'] = "Late Adults"
chest_metadata.loc[(chest_metadata['Patient Age'] >= 60), 'Patient Age Group'] = "Elderly"
```

```python
chest_metadata
```

```python
age_group = chest_metadata.groupby(['Patient Age Group'])
```

```python
# Put data in percentage
age_finding_distribution = []

# Get each age group and get their total and store the With Findings and No Findings Ratio in the dataframe above
for key, item in age_group:
    temp = age_group.get_group(key)['Finding Labels'].value_counts()
    temp_total = temp.sum()
    temp_no_findings_percent = temp['No Finding']/temp_total*100
    age_finding_distribution.append({'Age Group': key,
                                    'Percent No Finding': temp_no_findings_percent,
                                    'Percent with Findings': 100-temp_no_findings_percent})
    del temp, temp_total, temp_no_findings_percent
    
age_finding_distribution = pd.DataFrame(age_finding_distribution)
age_finding_distribution
```

```python
# Re-arrange Table
age_finding_distribution = age_finding_distribution.reindex([1, 5, 2, 0, 4, 3])
```

```python
age_finding_distribution['Percent with Findings'].plot(kind='bar', figsize=(10,5), rot=0)
plt.xlabel('Age Group', fontsize = 14)
plt.ylabel('Percentage', fontsize = 14)
plt.xticks(range(0,len(age_finding_distribution)), age_finding_distribution['Age Group'])
plt.title('Percentage of with Findings in Chest Radiographs', fontsize= 16)
plt.savefig('images_made/PercentFindingsAge.png', dpi=300, bbox_inches = 'tight')
```
